```python
def get_FASTA(text):
    result = []

    text_split = text.split('\n\n')
    
    for i in text_split:
        new_text = i.split('\n')
        id_cadena = re.findall(r"\|\d*\D*\d*\D*\|\D",new_text[0])[0].split('|')
        id_cadena = id_cadena[1] + "_" + id_cadena[2]
        FASTA = "".join([val for ind,val in enumerate(new_text) if ind != 0])
        dic = {}
        dic["id_cadena"] = id_cadena
        dic["FASTA"] = FASTA
        result.append(dic)
        
    return result
```
